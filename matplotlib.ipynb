```python
import numpy as np
import matplotlib.pyplot as plt
runs=np.array([100,50,91,78,89,25,34,19,9,10])#x-axis
w=np.array([1,0,2,0,3,7,8,9,7,5])#y-axis
plt.scatter(w,runs,color='blue')#scatter=graph name(x-axis,y-axis,color)
plt.title('IndvsAus_score')#to change the title
plt.legend(loc='best')#legend=to give best location
plt.show()#displaying the finalgraph
```

```python
#generate array of 200 values b/w -pi to pi
tigar=np.linspace(-2*np.pi,2*np.pi,100)
print(tigar)
```

```python
plt.plot(tigar,np.sin(tigar),color='black',linestyle='dotted')#(pit.plot(x,y,color,labels....))
plt.title("sin(x)")
plt.show()
```

```python
#plotting all in one graph
#creating x
overs = np.arange(5,50,5)
overs_a = np.arange(5,30,5)
#creating y
runs_i = np.array([25,51,84,131,160,189,220,250,267])
runs_a = np.array([15,41,94,110,151])
wickets = np.array([12,32,96])
#plotting
plt.plot(overs,runs_i,color='blue',label='India')#label=locate the best position for label
plt.plot(overs_a,runs_a,color='yellow',label = 'Aus')
plt.legend(loc='best')#combining two graphs
plt.show()
```

```python
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]#x-axis
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]#y-axis
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)#subplot=seperate graphs
plt.plot(years,profit_a,color='hotpink',linewidth= '7',label =#linewidth=to change line size
'CompanyA',marker='>',ms='20',mec='m')#ms=marker size,mec=marker outline color changes
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label =
'CompanyB',marker = 'H')
plt.show()
```

```python
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
col=['blue','green','pink','yellow']
plt.pie(a,labels = labe,colors=col)
plt.show()
```

```python
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
explo = [0,0.6,0,0]#cutting of particular part as seperate one
plt.pie(a,labels = labe,explode = explo,startangle =270)
plt.show()
```
